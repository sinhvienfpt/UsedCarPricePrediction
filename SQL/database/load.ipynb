```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv('../../assets/data//cleaned.csv')
df.head()
```

```python
# Load data into SQLite
conn = sqlite3.connect('old_car_db.db')
df.to_sql('train', conn, if_exists='replace', index=False)
```

```python
# Test with a query select the first 5 rows
query = """
SELECT * FROM train LIMIT 5
"""
pd.read_sql(query, conn)
```
